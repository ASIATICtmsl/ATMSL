```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download nltk data
nltk.download('punkt')
nltk.download('stopwords')

# Load your dataset
df = pd.read_csv('job_descriptions_with dept.csv', encoding='ISO-8859-1')

# Define keywords/phrases
keywords = [
    "Detail-oriented", "Time management", "Attention to detail", "Communication",
    "Communication skills", "Flexible", "Effective communicators", "Problem solving",
    "Collaborative", "Creativity", "Highly organized", "Organization", 
    "Organizational skills", "Trustworthiness", "Trustworthy", "Adaptability", 
    "Business", "Integrity", "Accountant", "Adaptable", "Analytical", 
    "Critical thinking", "Ethical", "Industry knowledge"
]

# Combine similar keywords into sets
keyword_sets = [
    {"Detail-oriented", "Attention to detail"},
    {"Communication", "Communication skills", "Effective communicators"},
    {"Organization", "Highly organized", "Organizational skills"},
    {"Trustworthiness", "Trustworthy"},
    {"Adaptability", "Adaptable"},
]

# Flatten keyword sets for simple search
all_keywords = set([item for sublist in keyword_sets for item in sublist] + [
    "Time management", "Flexible", "Problem solving", "Collaborative", 
    "Creativity", "Business", "Integrity", "Accountant", "Analytical", 
    "Critical thinking", "Ethical", "Industry knowledge"
])

# Compile a regex pattern for keywords
keyword_pattern = re.compile('|'.join([re.escape(keyword) for keyword in all_keywords]), re.IGNORECASE)

# Define a function to find matching sentences
def find_matching_sentences(text):
    sentences = nltk.sent_tokenize(text)
    matching_sentences = [sentence for sentence in sentences if any(keyword_pattern.search(word) for word in word_tokenize(sentence))]
    return ' '.join(matching_sentences)

# Apply the function only to rows where the job department is 'Accounts'
df.loc[df['Department'] == 'Accounts', 'Matching Sentences'] = df.loc[df['Department'] == 'Accounts', 'Requirements'].apply(find_matching_sentences)

# Save the DataFrame with the new column to a new CSV file
df.to_csv('job_descriptions_with.csv', index=False)
```
